```python
import matplotlib.pyplot as plt
metrics=[ 'Local Route', 'Light Rail', 'Peak Service', 'Rapid Route',
       'School']
for metric in metrics:
  data[metric].plot(figsize=(10,6),title=f"{metrics}trends")
  plt.show()
```

```python
from pmdarima import auto_arima
forecast_results={}
for metric in metrics:
  model=auto_arima(data[metric],seasonal=True,m=7)
  forecast=model.predict(n_periods=7)
  forecast_results[metric]=forecast
  print(f"Forecast for {metrics}:{forecast}")
```

```python
from prophet import Prophet
forecast_resuts={}
for metric in metrics:
  df=data[[metric]].reset_index().rename(columns={'Date':'ds', metric:'y'})
  model= Prophet()
  model.fit(df)
  future=model.make_future_dataframe(periods=7)
  forecast=model.predict(future)
  forecast_results[metric]=forecast[['ds','yhat']].tail(7)
  print(f"Forecast for {metric}:")
  print(forecast_results[metric])
```

```python
for metric in metrics:
  plt.figure(figsize=(10,6))
  plt.plot(data[metric],label='Historical Data')
  plt.plot(forecast_results[metric]['ds'],forecast_results[metric]['yhat'],label='Forecast')
  plt.title(f"{metric} Forecast")
  plt.legend()
  plt.show()
```
